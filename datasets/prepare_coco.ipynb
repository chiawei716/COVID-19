```python
import os
import cv2
import datetime
import json
```

```python
# Output JSON content
output = {
    "info": {
        "description": "COVID-19 X-Ray Dataset",#"url": "http://cocodataset.org",
        "version": "1.0",
        "year": 2020,
        "contributor": "TW",
        "date_created": "2020/06/26"
    },
    "categories": [
        {"id": 1, "name": "Pneumonia"},
        {"id": 2, "name": "GGO"},
        {"id": 3, "name": "LLL"}
        
    ],
    "licenses": [],
    "images": [],
    "annotations": []
}
```

```python
# Show dataset directory
%cd train_20200620
os.listdir()
```

```python
image_id = 1
annotation_id = 1
```

```python
# GGO
%cd "GGO PNG"
for filename in os.listdir():
    
    # Read image
    img = cv2.imread(filename)
    height, width, channel = img.shape
    
    # Create new image instance
    new_img = {
        #"license": 4,
        "file_name": "GGO PNG/" + filename,
        #"coco_url": "http://images.cocodataset.org/val2017/000000397133.jpg",
        "height": height,
        "width": width,
        "date_captured": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        #"flickr_url": "http://farm7.staticflickr.com/6116/6255196340_da26cf2c9e_z.jpg",
        "id": image_id
    }
    
    # Read label
    with open("../GGO Label/" + filename.split('.')[0] + "_01.txt", "r") as f:
        lines = f.readlines()
        
        # Remove '\n' & ','
        for i in range(len(lines)):
            lines[i] = lines[i].split('\n')[0]
            lines[i] = lines[i].split(',')
    
        # Create new annotation instance
        for i in range(int(len(lines) / 2)):
            new_annotation = {
                #"segmentation": [[510.66,423.01,511.72,420.03,...,510.45,423.01]],
                #"area": 702.1057499999998,
                "iscrowd": 0,
                "image_id": image_id,
                "bbox": [lines[i][0], lines[i][1], lines[i+1][0], lines[i+1][1]],
                "category_id": lines[i][2],
                "id": annotation_id
            }
            output['annotations'].append(new_annotation)
            annotation_id = annotation_id + 1
    
    output['images'].append(new_img)
    image_id = image_id + 1
    
%cd ..
```
