```python
import cv2
from matplotlib import pyplot as plt
```

## GGO

```python
!./darknet detector test data/COVID19.data cfg/tiny-try.cfg backup/tiny-try_best.weights 'data/COVID-19/GGO PNG/GGO021601.png'
```

```python
with open('../dataset/GGO Label/GGO021601_01.txt') as file:
    lines = file.readlines()
    for i in range(len(lines)):
        lines[i] = lines[i].split('\n')[0].split(',')
        
im = cv2.imread('data/COVID-19/GGO PNG/GGO021601.png')

for i in range(int(len(lines)/2)):
    if lines[i*2][2] == '1': # Pneumonia: Red
        color = (255, 0, 0)
    elif lines[i*2][2] == '2': # GGO: Green
        color = (0, 255, 0)
    elif lines[i*2][2] == '3': # LLL: Blue
        color = (0, 0, 255)
    cv2.rectangle(
        im, 
        (int(lines[i*2][0]), int(lines[i*2][1])), 
        (int(lines[i*2+1][0]), int(lines[i*2+1][1])), 
         color, 
         10
        )
plt.imshow(im)
```

```python
im = cv2.imread('./predictions.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
```
